```python
import pandas as pd


def label_data_based_on_keywords(df, column, keywords):

    if not isinstance(keywords, list):
        raise ValueError("Keywords must be provided as a list")


    if column not in df.columns:
        raise ValueError(f"Column '{column}' not found in DataFrame")


    df['type'] = df[column].apply(lambda x: 'non-veg' if any(keyword in str(x) for keyword in keywords) else 'veg')

    return df


keywords = ['chicken', 'beef', 'pork', 'lamb', 'turkey', 'duck', 'goat', 'rabbit', 'salmon', 'tuna', 'cod', 'haddock', 'shrimp', 'crab', 'lobster', 'oysters', 'squid', 'bacon', 'sausages', 'ham', 'salami', 'hot dogs', 'corned beef']


labeled_df = label_data_based_on_keywords(df, column='RecipeIngredientParts', keywords=keywords)


print(labeled_df)
```
